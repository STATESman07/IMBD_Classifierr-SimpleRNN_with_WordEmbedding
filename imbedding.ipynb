```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=["the glass of milk",
    "the glass of juice",
    "the cup of tea",
    "I am a good boy",
    "I am a good developer",
    "understand the meaning of words",
    "not understand the meaning of words",
    "your videos are good",
    ]
```

```python
sent
```

```python
for words in sent:
    print(words,sep=",")
```

```python
## defiing vocabulary size

## define vocab size
vocab_Size=10000
```

```python
### get one hot encoding
one_hot_Represntation=[one_hot(words,vocab_Size)for words in sent]
```

```python
sent
```

```python
one_hot_Represntation
```

```python
## word Embedding representation

from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

```python
## set max length
sent_length=8
embedding_doc=pad_sequences(one_hot_Represntation,padding='pre',maxlen=sent_length)
```

```python
embedding_doc
```

```python
# feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(input_dim=vocab_Size,
                    output_dim=dim,
                    input_length=sent_length))
model.build(input_shape=(None,sent_length))
model.compile('adam','mse')

```

```python
model.summary()
```

```python
## converted embeddind doc into vector


model.predict(embedding_doc)
```
